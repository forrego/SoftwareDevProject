```python
import pandas as pd
```

```python
stop_data1_df = pd.read_csv("2018-2019 Stop Data PART 1.csv")
stop_data2_df = pd.read_csv("2018-2019 Stop Data PART 2.csv")
```

```python
merged_df = [stop_data1_df, stop_data2_df]
stop_data_df = pd.concat(merged_df)
stop_data_df
```

```python
stop_data_df.info()
```

```python
# extracting silver route
silver_df = stop_data_df[stop_data_df['Route'] == 'Green']
silver_df
```

```python
# getting applicable columns
filtered_df = silver_df[['Date', 'Count', 'Onoff', 'Stop']]
filtered_df
```

```python
# dropping 
filtered_df.dropna()
filtered_df.describe()
```

```python
filtered_df.Date.unique().size
```

```python
on_df = filtered_df[filtered_df['Onoff'] == 'on']
off_df = filtered_df[filtered_df['Onoff'] == 'off']
```

```python
on_by_stop = on_df.groupby('Stop')['Count'].sum()
off_by_stop = off_df.groupby('Stop')['Count'].sum()
```

```python
on_by_stop.to_csv(r'green_on.csv')
on_by_stop
```

```python
off_by_stop.to_csv(r'green_off.csv')
```
